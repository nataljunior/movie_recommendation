```python
# keeping only title, instead of original_title : Porco Rosso | 紅の豚
keep1 = ['title', 'text', 'release_date', 'genres',
         'status', 'vote_average', 'vote_count',
         'runtime', 'revenue', 'budget', 'original_language', 'popularity']

df1 = df1[keep1]
```

```python
df1 = df1.dropna(subset=['title'])
```

```python
#from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_extraction.text import CountVectorizer # convert into vector
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
```

```python
import ipywidgets as widgets
from IPython.display import display
```

```python
# to initiate the class, we need to set the language (bag of words) and max_features = number of columns
```

```python
cv = CountVectorizer(max_features=len(df1), stop_words='english')
```

```python
cv
```

```python
vec =  cv.fit_transform(df1['text'].values.astype('U')).toarray() # inst run 10k | 7s run 50k |100000 ram prob
```

```python
vec.shape
```

```python
# create the recommendation system
cosim = cosine_similarity(vec)
```

```python
cosim
```

```python
def recommand(movies):
    index = df1[df1['title'] == movies].index[0]
    distance = sorted(list(enumerate(cosim[index])), reverse=True, key=lambda vec: vec[1])
    for i in distance[0:5]:
        movie_info = df1.iloc[i[0]]
        print(f"Title: {movie_info.title}, Vote Average: {movie_info.vote_average}, Release Date: {movie_info.release_date}, Popularity: {movie_info.popularity}")
```

```python
from tabulate import tabulate

def recommand(movies):
    index = df1[df1['title'] == movies].index[0]
    distance = sorted(list(enumerate(cosim[index])), reverse=True, key=lambda vec: vec[1])

    recommended_movies = []
    for i in distance[1:6]:
        movie_info = df1.iloc[i[0]]
        recommended_movies.append([movie_info.title, movie_info.vote_average, movie_info.release_date, movie_info.popularity])

    headers = ["Title", "Vote Average", "Release Date", "Popularity"]
    print(tabulate(recommended_movies, headers=headers, tablefmt="pretty"))
```

```python
import ipywidgets as widgets
from IPython.display import display, HTML
from tabulate import tabulate

movie_name_input = widgets.Text(
    value='Insert Movie Name',
    description='Movie Title:',
    disabled=False
)

recommendation_table = widgets.Output()

def recommand(movies):
    matching_movies = df1[df1['title'] == movies]
    if not matching_movies.empty:
        index = matching_movies.index[0]
        distance = sorted(list(enumerate(cosim[index])), reverse=True, key=lambda vec: vec[1])

        recommended_movies = []
        for i in distance[1:6]: # using 0:5 will show the movie searched as first
            movie_info = df1.iloc[i[0]]
            recommended_movies.append([movie_info.title, movie_info.vote_average, movie_info.release_date, movie_info.popularity])

        headers = ["Title", "Vote Average", "Release Date", "Popularity"]
        with recommendation_table:
            recommendation_table.clear_output()
            print(tabulate(recommended_movies, headers=headers, tablefmt="pretty"))
    else:
        with recommendation_table:
            recommendation_table.clear_output()
            print("No matching movies found.")

def on_type(data):
    with recommendation_table:
        recommendation_table.clear_output()
        title = data["new"]
        if len(title) > 0:
            recommand(title)

movie_name_input.observe(on_type, names='value')

display(movie_name_input, recommendation_table)
```
